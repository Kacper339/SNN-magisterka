```python
import torch
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from pathlib import Path
from matplotlib.backends.backend_pdf import PdfPages
from sklearn.metrics import classification_report, confusion_matrix, top_k_accuracy_score
import torch.nn as nn
import math
import torch.nn.functional as F
from torch.utils.data import Dataset, DataLoader

# =========================
# ŚCIEŻKI
# =========================
CHECKPOINT_PATH = Path(r"C:\Users\Kacper\OneDrive\Desktop\SNN\results\results-real_101- VGG Net\checkpoint.pth")
BEST_MODEL_PATH = Path(r"C:\Users\Kacper\OneDrive\Desktop\SNN\results\results-real_101- VGG Net\best_model.pth")
DATA_DIR = Path(r"C:\Users\Kacper\OneDrive\Desktop\SNN\data\N-Caltech101-full-split")
EXPERIMENT_DIR = Path(r"C:\Users\Kacper\OneDrive\Desktop\SNN\plots\results-real_101- VGG Net")
ANALYSIS_DIR = EXPERIMENT_DIR / "analysis"
ANALYSIS_DIR.mkdir(parents=True, exist_ok=True)

print("Wyniki zapisują się w:", ANALYSIS_DIR)

def save_figure(fig, name):
    fig.savefig(ANALYSIS_DIR / f"{name}.png", dpi=300, bbox_inches="tight")
    plt.close(fig)

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# =========================
# FUNKCJE POMOCNICZE
# =========================
def load_bin_events(path):
    raw = np.fromfile(path, dtype=np.uint8)
    raw = raw[:len(raw) // 5 * 5]
    x = raw[0::5].astype(np.int32)
    y = raw[1::5].astype(np.int32)
    p = (raw[2::5] >> 7) & 1
    p = np.where(p == 1, 1, -1)
    t_high = (raw[2::5] & 0x7F).astype(np.uint32)
    t_mid  = raw[3::5].astype(np.uint32)
    t_low  = raw[4::5].astype(np.uint32)
    t = (t_high << 16) | (t_mid << 8) | t_low
    return x, y, t.astype(np.float32), p

class EventDatasetBIN(Dataset):
    def __init__(self, root_dir, classes, T=10, H=180, W=240, clip_value=5.0, augment=False):
        self.root_dir = Path(root_dir)
        self.classes = classes
        self.class_to_idx = {c: i for i, c in enumerate(classes)}
        self.T, self.H, self.W = T, H, W
        self.clip_value = clip_value
        self.augment = augment
        self.samples = []
        for cls in classes:
            for f in (self.root_dir / cls).glob("*.bin"):
                self.samples.append((f, self.class_to_idx[cls]))

    def __len__(self):
        return len(self.samples)

    def __getitem__(self, idx):
        path, label = self.samples[idx]
        x, y, t, p = load_bin_events(path)
        if self.augment and np.random.rand() > 0.5:
            x = (self.W - 1) - x
        t = (t - t.min()) / (t.max() - t.min() + 1e-9)
        voxels = np.zeros((2, self.H, self.W, self.T), dtype=np.float32)
        time_bins = (t * (self.T - 1)).astype(np.int32)
        for xi, yi, pi, bi in zip(x, y, p, time_bins):
            if 0 <= xi < self.W and 0 <= yi < self.H:
                ch = 0 if pi > 0 else 1
                voxels[ch, yi, xi, bi] += 1.0
        voxels = np.clip(voxels, 0.0, self.clip_value)
        voxels /= self.clip_value
        return torch.from_numpy(voxels), torch.tensor(label)

# =========================
# WŁADANIE CHECKPOINTU HISTORY
# =========================
checkpoint = torch.load(CHECKPOINT_PATH, map_location=device, weights_only=False)
history = checkpoint["history"]

epochs = history["epoch"]
train_loss = history["train_loss"]
val_loss   = history["val_loss"]
train_acc  = history["train_acc"]
val_acc    = history["val_acc"]
top1       = history["top1"]
top5       = history["top5"]
spike_rates = np.array(history["spike_rate"])

print(f"Liczba epok: {len(epochs)}")
print(f"Najlepsza val_acc: {checkpoint['best_val_acc']*100:.2f}%")

# =========================
# WYKRESY HISTORY
# =========================
print("Tworzenie wykresów historycznych...")
# Loss
fig = plt.figure()
plt.plot(epochs, train_loss, label="Train Loss")
plt.plot(epochs, val_loss, label="Val Loss")
plt.xlabel("Epoka"); plt.ylabel("Loss"); plt.title("Loss podczas treningu"); plt.legend(); plt.grid(True)
save_figure(fig, "loss_curve")
# Accuracy
fig = plt.figure()
plt.plot(epochs, train_acc, label="Train Accuracy")
plt.plot(epochs, val_acc, label="Val Accuracy")
plt.xlabel("Epoka"); plt.ylabel("Accuracy"); plt.title("Accuracy podczas treningu"); plt.legend(); plt.grid(True)
save_figure(fig, "accuracy_curve")
# Top1 / Top5
fig = plt.figure()
plt.plot(epochs, np.array(top1)*100, label="Top-1 Acc")
plt.plot(epochs, np.array(top5)*100, label="Top-5 Acc")
plt.xlabel("Epoka"); plt.ylabel("Accuracy [%]"); plt.title("Top-1 / Top-5 Accuracy"); plt.legend(); plt.grid(True)
save_figure(fig, "top1_top5")
# Generalization Gap
gap = np.array(train_acc) - np.array(val_acc)
fig = plt.figure()
plt.plot(epochs, gap)
plt.xlabel("Epoka"); plt.ylabel("Train - Val Accuracy"); plt.title("Generalization Gap"); plt.grid(True)
save_figure(fig, "generalization_gap")
# Spike Rate per Layer
fig = plt.figure()
for i in range(spike_rates.shape[1]):
    plt.plot(epochs, spike_rates[:, i], label=f"LIF{i+1}")
plt.xlabel("Epoka"); plt.ylabel("Średnia częstość spikowania"); plt.title("Spike rate per warstwa"); plt.legend(); plt.grid(True)
save_figure(fig, "spike_per_layer")
# Spike Rate vs Val Accuracy
mean_spike = spike_rates.mean(axis=1)
fig = plt.figure()
plt.scatter(mean_spike, val_acc)
plt.xlabel("Średni Spike Rate"); plt.ylabel("Validation Accuracy"); plt.title("Spike Rate vs Val Accuracy"); plt.grid(True)
save_figure(fig, "spike_vs_val_acc")
corr = np.corrcoef(mean_spike, val_acc)[0,1]
print(f"Korelacja Pearson Spike Rate vs Val Accuracy: {corr:.4f}")
# Spike stability
std_spike = spike_rates.std(axis=1)
fig = plt.figure()
plt.plot(epochs, std_spike)
plt.xlabel("Epoka"); plt.ylabel("Odchylenie standardowe spike rate"); plt.title("Stabilność aktywności sieci"); plt.grid(True)
save_figure(fig, "spike_stability")

# =========================
# ZAPIS METRICS CSV
# =========================
metrics_df = pd.DataFrame({
    "epoch": epochs,
    "train_loss": train_loss,
    "val_loss": val_loss,
    "train_acc": train_acc,
    "val_acc": val_acc,
    "mean_spike_rate": mean_spike,
    "spike_std": std_spike
})
metrics_df.to_csv(ANALYSIS_DIR / "metrics.csv", index=False)
print("metrics.csv zapisany ✅")

# =========================
# WŁADANIE MODELU (CPU) + MACIERZ POMYŁEK
# =========================
print("Ładowanie modelu i ewaluacja na CPU...")
# Tu zakładamy, że masz funkcję vgg11() jak wcześniej wklejona
# i listę klas CLASSES (z nazw folderów)
CLASSES = sorted([f.name for f in (DATA_DIR / "test").iterdir() if f.is_dir()])
test_dataset = EventDatasetBIN(DATA_DIR / "test", CLASSES)
test_loader = DataLoader(test_dataset, batch_size=1, shuffle=False)

model = vgg11(num_classes=len(CLASSES), in_c=2, batch_norm=True,
              thresh=1.35, tau=0.9, gamma=1.0, dropout_p=0.3).to("cpu")
model.load_state_dict(torch.load(BEST_MODEL_PATH, map_location="cpu", weights_only=False))
model.eval()

all_labels, all_probs = [], []

criterion = nn.CrossEntropyLoss()
with torch.no_grad():
    for i, (x, y) in enumerate(test_loader):
        x = x.permute(0,4,1,2,3)  # (B,C,H,W,T) -> (B,T,C,H,W)
        out = model(x)
        out = out.mean(1)
        probs = F.softmax(out, dim=1)
        all_labels.extend(y.numpy())
        all_probs.extend(probs.numpy())
        if (i+1) % 50 == 0:
            print(f"Przetworzono {i+1}/{len(test_loader)} próbek...")

all_labels = np.array(all_labels)
all_probs = np.array(all_probs)
preds = np.argmax(all_probs, axis=1)

# =========================
# RAPORT KLASYFIKACJI
# =========================
report_txt = classification_report(all_labels, preds, target_names=CLASSES)
with open(ANALYSIS_DIR / "classification_report.txt", "w", encoding="utf-8") as f:
    f.write("=== Classification Report ===\n")
    f.write(report_txt)
    f.write(f"\nPearson Spike/Val Acc Correlation: {corr:.4f}\n")
print("Raport klasyfikacji zapisany ✅")

# =========================
# MACIERZ POMYŁEK
# =========================
cm = confusion_matrix(all_labels, preds)
fig = plt.figure(figsize=(12,10))
sns.heatmap(cm, annot=False, cmap="Blues")
plt.title("Macierz pomyłek")
plt.xlabel("Predykcja")
plt.ylabel("Rzeczywista")
save_figure(fig, "confusion_matrix")
print("Macierz pomyłek zapisna ✅")

# =========================
# RAPORT PDF Z PLIKÓW PNG
# =========================
report_path = ANALYSIS_DIR / "raport_eksperymentu.pdf"
with PdfPages(report_path) as pdf:
    for file in ANALYSIS_DIR.glob("*.png"):
        img = plt.imread(file)
        fig = plt.figure(figsize=(8,6))
        plt.imshow(img)
        plt.axis("off")
        pdf.savefig(fig)
        plt.close(fig)

print("Raport PDF gotowy ✅")
print("Analiza zakończona. Wszystkie pliki w:", ANALYSIS_DIR)
```
