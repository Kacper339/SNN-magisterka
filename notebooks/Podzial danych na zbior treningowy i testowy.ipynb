```python
import os
import random
import shutil
from pathlib import Path

# ============================================================
# PARAMETRY
# ============================================================

DATA_ROOT = Path(r"C:\Users\Kacper\OneDrive\Desktop\SNN\data\Synthetic_DVS")
OUT_ROOT  = Path(r"C:\Users\Kacper\OneDrive\Desktop\SNN\data\Synthetic_DVS-full-split")

EXTENSIONS = [".h5"]   # wybierze tylko te pliki
TRAIN_RATIO = 0.8
SEED = 42

random.seed(SEED)

# ============================================================
# FUNKCJA PODZIAŁU JEDNEJ KLASY
# ============================================================

def split_class(class_dir: Path, out_root: Path):
    class_name = class_dir.name
    
    files = [
        f for f in class_dir.iterdir()
        if f.is_file() and f.suffix.lower() in EXTENSIONS
    ]
    
    if len(files) < 5:
        print(f"⚠️ Pomijam klasę {class_name} (za mało plików)")
        return
    
    random.shuffle(files)
    
    n_train = int(len(files) * TRAIN_RATIO)
    train_files = files[:n_train]
    test_files  = files[n_train:]
    
    for split, split_files in [("train", train_files), ("test", test_files)]:
        target_dir = out_root / split / class_name
        target_dir.mkdir(parents=True, exist_ok=True)
        
        for f in split_files:
            shutil.copy(f, target_dir / f.name)

    print(
        f"✔ {class_name:<10s} | total={len(files):3d} | "
        f"train={len(train_files):3d} | test={len(test_files):3d}"
    )

# ============================================================
# GŁÓWNA PĘTLA
# ============================================================

def split_dataset():
    print("📂 Dzielenie zbioru danych...\n")
    
    for class_dir in sorted(DATA_ROOT.iterdir()):
        if class_dir.is_dir():
            split_class(class_dir, OUT_ROOT)
    
    print("\n✅ Podział zakończony")

# ============================================================
# START
# ============================================================

if __name__ == "__main__":
    split_dataset()
```
